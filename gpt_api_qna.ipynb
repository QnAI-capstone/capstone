```python
from google.colab import drive
import os
import json

# JSON 파일 경로 지정
json_dir = '/content/drive/MyDrive/jiwoo/data/json'

for file in os.listdir(json_dir):
    if file.endswith(".json"):
        path = os.path.join(json_dir, file)
        try:
            with open(path, 'r', encoding='utf-8') as f:
                json.load(f)
        except json.JSONDecodeError as e:
            print(f"\n❌ 오류 발생: {file}")
            print(f"📍 줄 번호: {e.lineno}, 열: {e.colno}")

            # 오류 발생한 줄의 내용을 함께 출력
            with open(path, 'r', encoding='utf-8') as f:
                lines = f.readlines()
                error_line = lines[e.lineno - 1].strip() if e.lineno <= len(lines) else "<줄 없음>"
                print(f"🔍 해당 줄 내용:\n{error_line}")
```

```python
# JSON 파일 경로 지정
json_dir = '/content/drive/MyDrive/jiwoo/data/json'

# JSON 파일들 모두 불러오기
data_list = []
for file in os.listdir(json_dir):
    if file.endswith(".json"):
        with open(os.path.join(json_dir, file), 'r', encoding='utf-8') as f:
            data_list.append(json.load(f))
```

```python
import openai
from dotenv import load_dotenv

# .env 파일 경로 지정
env_path = '/content/drive/MyDrive/jiwoo/.env'

# .env 파일 로드
load_dotenv(dotenv_path = env_path)

##### 수정필요
openai.api_key = os.getenv("OPENAI_API_KEY")

# JSON들을 하나의 문자열로 통합 (너무 길면 요약 필요)
context = "\n\n".join([json.dumps(d, ensure_ascii=False) for d in data_list])

# 질문 입력
user_question = input("질문을 입력하세요: ")

# GPT 호출
response = openai.ChatCompletion.create(
    model="gpt-4o",
    messages=[
        {"role": "system", "content": "너는 JSON 데이터를 기반으로 질문에 답하는 도우미야."},
        {"role": "user", "content": f"다음 데이터를 참고해서 질문에 답해줘:\n{context}"},
        {"role": "user", "content": user_question}
    ],
    temperature=0.2
)

# 결과 출력
print(response['choices'][0]['message']['content'])
```
