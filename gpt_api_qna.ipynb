```python
# JSON íŒŒì¼ ê²½ë¡œ ì§€ì •
json_dir = '/content/drive/MyDrive/jiwoo/data/json'

# JSON íŒŒì¼ë“¤ ëª¨ë‘ ë¶ˆëŸ¬ì˜¤ê¸°
data_list = []
for file in os.listdir(json_dir):
    if file.endswith(".json"):
        with open(os.path.join(json_dir, file), 'r', encoding='utf-8') as f:
            data_list.append(json.load(f))
```

```python
import openai
from dotenv import load_dotenv

# .env íŒŒì¼ ê²½ë¡œ ì§€ì •
env_path = '/content/drive/MyDrive/jiwoo/.env'

# .env íŒŒì¼ ë¡œë“œ
load_dotenv(dotenv_path = env_path)

##### ìˆ˜ì •í•„ìš”
openai.api_key = os.getenv("OPENAI_API_KEY")

# JSONë“¤ì„ í•˜ë‚˜ì˜ ë¬¸ìì—´ë¡œ í†µí•© (ë„ˆë¬´ ê¸¸ë©´ ìš”ì•½ í•„ìš”)
context = "\n\n".join([json.dumps(d, ensure_ascii=False) for d in data_list])

# ì§ˆë¬¸ ì…ë ¥
user_question = input("ì§ˆë¬¸ì„ ì…ë ¥í•˜ì„¸ìš”: ")

# GPT í˜¸ì¶œ
response = openai.ChatCompletion.create(
    model="gpt-4o",
    messages=[
        {"role": "system", "content": "ë„ˆëŠ” JSON ë°ì´í„°ë¥¼ ê¸°ë°˜ìœ¼ë¡œ ì§ˆë¬¸ì— ë‹µí•˜ëŠ” ë„ìš°ë¯¸ì•¼."},
        {"role": "user", "content": f"ë‹¤ìŒ ë°ì´í„°ë¥¼ ì°¸ê³ í•´ì„œ ì§ˆë¬¸ì— ë‹µí•´ì¤˜:\n{context}"},
        {"role": "user", "content": user_question}
    ],
    temperature=0.2
)

# ê²°ê³¼ ì¶œë ¥
print(response['choices'][0]['message']['content'])
```
